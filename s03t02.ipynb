### Exercici 3
Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
mat = vec.reshape(5,2)
newMat = np.array([[30, 40, 50], [50, 60, 70], [70, 80, 90], [90, 100, 110], [110, 120, 130]])
mat = np.append(mat, newMat, axis = 1)
print(mat)
# View or copy?
print("Max value in the matrix: ", np.amax(mat))
print("Max value in horizontal axis: ", mat.max(0))
print("Max value in vertical axis: ", mat.max(1))

# print(vec.reshape(5, 2).base) # returns the original array, so it's a view
```
